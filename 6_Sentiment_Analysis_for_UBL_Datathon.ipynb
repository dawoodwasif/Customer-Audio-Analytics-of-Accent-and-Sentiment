### Create initializer

```python
from pysentimiento import create_analyzer

"""
Emotion Analysis in English
"""

sentiment_analyzer = create_analyzer(task="emotion", lang="en")

```

### Inference

```python
sentiment = sentiment_analyzer.predict("I am very satisfied by your customer experience")
sentiment
```

```python
stats = sentiment.probas
main_sentiment = max(stats, key=stats.get)
perc = 0
for sen, per in stats.items():
    if sen == main_sentiment:
        perc = per * 100
if main_sentiment == 'other':
  main_sentiment = 'neutral'
print( 'User feels ' + main_sentiment + ' with ' + str(perc) + '% accuracy according to his words.')
```
