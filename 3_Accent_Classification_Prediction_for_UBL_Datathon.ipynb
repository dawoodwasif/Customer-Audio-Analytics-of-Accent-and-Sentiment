```python
import librosa
import numpy as np
import pandas as pd
```

### Extract input audio features

```python
path = '/content/farsi1.mp3'
data, sampling_rate = librosa.load(path)
```

```python
result = np.array([])
mel = np.mean(librosa.feature.melspectrogram(y=data, sr=sampling_rate).T, axis=0)
result = np.hstack((result, mel))
l = [77, 80, 65, 72, 63, 52, 87, 76]
l_features = (result[l]).tolist()
col_labels = ['mel77', 'mel80', 'mel65', 'mel72', 'mel63', 'mel52', 'mel87', 'mel76']
df = pd.DataFrame({
    col_labels[0]:[l_features[0]],
    col_labels[1]:[l_features[1]],
    col_labels[2]:[l_features[2]],
    col_labels[3]:[l_features[3]],
    col_labels[4]:[l_features[4]],
    col_labels[5]:[l_features[5]],
    col_labels[6]:[l_features[6]],
    col_labels[7]:[l_features[7]]

})
df.to_csv('/content/feat_pred.csv', index=False)
df.head()
```

### Export MOJO File and predict

```python
import h2o
h2o.init()

mojo_model = h2o.upload_mojo('/content/GBM_1_AutoML_4_20220401_125304.zip')
```

```python
new_observations = h2o.import_file(path='/content/feat_pred.csv')
predictions = mojo_model.predict(new_observations)
predictions
```
