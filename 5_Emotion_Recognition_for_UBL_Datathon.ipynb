```python
from pydub import AudioSegment
```

### Convert mp3 to wav file

```python
# files                                                                         
src = '/content/farsi1.mp3'
dst = '/content/test1.wav'

# convert wav to mp3                                                            
sound = AudioSegment.from_mp3(src)
sound.export(dst, format="wav")
```

### Initialize and train model

```python
from deep_emotion_recognition import DeepEmotionRecognizer
# initialize instance
# inherited from emotion_recognition.EmotionRecognizer
# default parameters (LSTM: 128x2, Dense:128x2)
deeprec = DeepEmotionRecognizer(emotions=['angry', 'sad', 'neutral', 'ps', 'happy'], n_rnn_layers=2, n_dense_layers=2, rnn_units=128, dense_units=128)
```

```python
# train the model
deeprec.train()
# get the accuracy
print(deeprec.test_score())
```

### Inference

```python
#deep_rec = tf.saved_model.load('/content/emotion-recognition-using-speech/results/AHNPS-c-LSTM-layers-2-2-units-128-128-dropout-0.3_0.3_0.3_0.3.h5')
```

```python
# predict angry audio sample
prediction = deeprec.predict(dst)
print(f"Prediction: {prediction}")
```

```python
emo_stats = deeprec.predict_proba(dst)
main_emotion = max(emo_stats, key=emo_stats.get)
perc = 0
for emt, per in emo_stats.items():
    if emt == main_emotion:
        perc = per * 100
if main_emotion == 'ps':
  main_emotion = 'pleasant or suprised'
print( 'User is ' + main_emotion + ' with a ' + str(perc) + '% accuracy according to his voice.')
```
